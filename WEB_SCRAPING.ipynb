```python
title = []
release_year = []
description = []
director = []
production = []
cast = []
genre = [] 
url = []
for j in tqdm ( movies_list['url'], desc="Loading…",ascii=False, ncols=75):
    time.sleep(0.01)
    
    try:
        webpage = requests.get(j).text
        soup1=BeautifulSoup(webpage,"html.parser")
    except:
        print('could not found url')
        continue
    # title
    try:
        title.append(soup1.find('h1').text)
    except:
        title.append(np.nan)
        
    # description
    try:
        description.append(soup1.find('div', class_='desc shorting').text)
    except:
        description.append(np.nan)
        
    #release year
    try:
        soup1.find_all('span')[16].text
        release_year.append(soup1.find_all('span')[16].text)
    except:
        release_year.append(np.nan)
        
    #Director
    try:
        director.append(soup1.find('span', class_='shorting').text)
    except:
        director.append(np.nan)
        
    # production
    try:
        production.append(soup1.find_all('span')[20].text)
    except:
        production.append(np.nan)
        
    #cast
    try:
        cast.append(soup1.find('div', class_= 'casts').text[6:].strip())
    except:
        cast.append(np.nan)
        
    # genre
    try:
        genre.append(soup1.find_all('span')[14].text)
    except:
        genre.append(np.nan)
    
    try:
        url.append(j)
    except:
        url.append(np.nan)

print("Complete.")

movies_detail = {'title':title, 'release_year':release_year, 'description':description,
                 'director':director, 'production':production,'cast':cast, 'genre':genre, 'URL':url}   
df_movies_detail = pd.DataFrame(movies_detail)
df_movies_detail
```
